### 상품명 오류 수정(문자열 전처리)

### item_price의 결측치 처리


```python
# 이번엔 uriage데이터의 item_price에 있던 NaN을 처리
# 결측치가 있는지 알아보기
uriage.isnull().any(axis=0) #item_price에만 있음
uriage.isnull().sum()
```

```python
# 고유한 상품명 26개를 for문으로 돌려 상품명 하나에 해달할 때마다 전체데이터프레임에서 NaN값을 올바른 가격값으로 채우기
for i in list(uriage['item_name'].unique()):
    print(i, '최저가:',uriage.loc[uriage['item_name'] == i, 'item_price'].min(skipna=False), '최고가:', uriage.loc[uriage['item_name'] == i, 'item_name'].max())#결측치가 있으니 NaN으로 나옴
# df.loc[행 인덱스, 열 인덱스] uriage.loc[uriage['item_name'] == i, 'item_price'] : uriage['item_name']이 상품A(예를들어)에 해당하는 item_price열 값
```

```python
flg_is_null = uriage['item_price'].isnull() #price가 NaN인 애들만 뽑음 T/F로 되어있으니 인덱싱 True인 애들만 뽑음
flg_is_null
```

```python
# flg_is_null = False일 때 item_price값이 있는 것이므로 False값을 찾음
for i in list(uriage['item_name'].unique()):
    price = uriage.loc[(~flg_is_null) & (uriage['item_name'] == i), 'item_price'].max()
    uriage.loc[(flg_is_null) & (uriage['item_name'] == i), 'item_price'] = price
uriage.head()
# price에 ~flg_is_null = (false, 값이 있는 것) 과 uriage[item_name]인 것에 해당하는 item_price의 최대값
# flg_is_null = (True, 값이 없는 것에) price를 넣음
# 상품A,B,C 등등이 위에서 통일 된 후 NaN이 있을 수 있으므로 상품A의 값을 바뀐 상품A에 값에 넣어줌
```

### 엑셀 날짜 오류 수정

```python
ko.dtypes
```

```python
ko['등록일'].astype('str').str.isdigit() # 숫자가 있는지 찾기 isdigit()
```

```python
ko['등록일'].astype('str').str.isdigit().sum() # 잘못 입력된 것의 개수 22개
```

```python
# 숫자가 들어가있는 데이터를 float타입으로 변형한 뒤 형식을 D(day)로 바꾸고 1900-01-01 날짜 더하기
# 먼저 숫자이면 True, 아니면 False인 불린 변수로 만든다.
is_serial = ko['등록일'].astype('str').str.isdigit()

#위의 변수를 불린 인덱싱 하여, 날짜인데 숫자로 잘못 들어간 애들을 날짜 단위로 변환
pd.to_timedelta(ko.loc[is_serial, '등록일'].astype('float')-2, unit='D')
# ko['등록일']에서 숫자로 되어있는 것을 float로 바꾸고 형식을 Day로 바꾼다 - timedelta를 이용해
# 엑셀 데이터는 0이아닌 1부터 시작 파이썬은 0부터 시작
# 액셀의 숫자형식은 파이썬으로 변환 시 이틀차이 그래서 -2해야함
```

```python
from_serial = pd.to_timedelta(ko.loc[is_serial, '등록일'].astype('float')-2, unit='D') + pd.to_datetime("1900/01/01")
from_serial
```

```python
#앞서 숫자로 판명되지 않은 것 (False)도 datetime으로 바꿔주기
from_string = pd.to_datetime(ko.loc[~is_serial, '등록일'])

# 두 변수(from_serial, from_string) 조인하여 데이터 프레임에 넣기 (세로로 붙이기)
ko['등록일'] = pd.concat([from_serial, from_string])
ko.head()
```
